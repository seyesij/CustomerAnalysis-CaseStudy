```python
#Read the three files into python as dataframes

import pandas as pd
import numpy as np

file1 = pd.read_csv('file1.csv')
file2 = pd.read_csv('file2.csv')
file3 = pd.read_csv('file3.csv')
```

```python
#Show the DataFrame's shape

print(file1.shape)
print(file2.shape)
print(file3.shape)
```

```python
#Standardize header names.

#columns to lowecase
cols1 = [column.lower() for column in file1.columns]
cols2 = [column.lower() for column in file2.columns]
cols3 = [column.lower() for column in file3.columns]
file1.columns = cols1
file2.columns = cols2
file3.columns = cols3

#renaming columns
file1 = file1.rename(columns={'st':'state'})
file2 = file2.rename(columns={'st':'state'})
```

```python
#Rearrange the columns in the dataframe as needed

file1 = file1[['customer', 'state', 'gender', 'education', 'customer lifetime value', 'income', 'policy type', 'monthly premium auto', 'number of open complaints', 'vehicle class', 'total claim amount']]
file2 = file2[['customer', 'state', 'gender', 'education', 'customer lifetime value', 'income', 'policy type', 'monthly premium auto', 'number of open complaints', 'vehicle class', 'total claim amount']]
file3 = file3[['customer', 'state', 'gender', 'education', 'customer lifetime value', 'income', 'policy type', 'monthly premium auto', 'number of open complaints', 'vehicle class', 'total claim amount']]
```

```python
print(file1.columns)
print(file2.columns)
print(file3.columns)
```

```python
#Correct the values in the column customer lifetime value

file1['customer lifetime value'] = file1['customer lifetime value'].str.rstrip('%').astype('float')/100.0 #Thanks Max!
file2['customer lifetime value'] = file2['customer lifetime value'].str.rstrip('%').astype('float')/100.0

#file1['customer lifetime value']
file2['customer lifetime value']
```

```python
#Concatenate the three dataframes

data = pd.concat([file1, file2, file3], axis=0) 
data.shape
data
```

```python
#Which columns are numerical?

numcols = data._get_numeric_data().columns.tolist()
print(numcols)
```

```python
#Which columns are categorical?

catcols = data.select_dtypes('object').columns.tolist()
catcols
```

```python
#Understand the meaning of all columns


data.dtypes
```

```python
#Perform the data cleaning operations mentioned so far in class

#Delete the column education and the number of open complaints from the dataframe.
data = data.drop(['education', 'number of open complaints'], axis=1)

data
```

```python
print(data.shape)
```

```python
#Check for duplicate rows in the data and remove if any.

data = data.drop_duplicates()
print(data.shape)
```

```python
#Filter out the data for customers who have an income of 0 or less.

filtered = data[data['income']>0]
filtered
```
